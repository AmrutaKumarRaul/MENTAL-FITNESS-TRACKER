```python
# Import the required libraries and mount Google Drive.
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df1 = pd.read_csv('/content/drive/MyDrive/datasheet/mental-and-substance-use-as-share-of-disease -AI.csv')
df2 = pd.read_csv('/content/drive/MyDrive/datasheet/prevalence-by-mental-and-substance-use-disorder _AI.csv')
```

```python
# Display the first few rows of df1 using the 'head()' function
print("Mental Health Data:")
print(df1.head())

print("\nMental Disorder Data:")
print(df2.head())
```

```python
# Check the data types and null values, if any.
print("\nData Types:")
print(df1.dtypes)
print("\nNull Values:")
print(df1.isnull().sum())
```

```python

# Plotting a simple line chart to visualize the trend of 'Prevalence - Depressive disorders'
plt.figure(figsize=(10, 6))
plt.plot(df2['Year'], df2['Prevalence - Depressive disorders - Sex: Both - Age: Age-standardized (Percent)'])
plt.xlabel('Year')
plt.ylabel('Prevalence - Depressive disorders')
plt.title('Trend of Prevalence - Depressive disorders over the years')
plt.grid(True)
plt.show()
```

```python
print("Columns in df1:")
print(df1.columns)

print("\nColumns in df2:")
print(df2.columns)
```

```python
# Select features from the dataset.
selected_features_df = df2[['Year', 'Prevalence - Depressive disorders - Sex: Both - Age: Age-standardized (Percent)']]
target_df = df1[['DALYs (Disability-Adjusted Life Years) - Mental disorders - Sex: Both - Age: All Ages (Percent)']]
```

```python
# Split the dataset into training and testing sets.
X_train, X_test, y_train, y_test = train_test_split(selected_features_df, target_df, test_size=0.2, random_state=42)

# Initialize and fit the linear regression model.
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Predict using the model.
y_pred = model.predict(X_test)

# Evaluate the model.
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
print(f"Mean Squared Error: {mse}")
print(f"R-squared Score: {r2}")
```
